```python
# pre uses the pivot module to create the header based on the header function from cvslist. The pivot function takes 
# the first two column headers and adds "Column" and "Value". The preexisting.data allows for each of the first two 
# columns to be repeated for each month and amount per month. The list(range(2,37)) is what tells the pivot function 
# to pivot columns 0 through 37.

# pre_clean takes the first row of pre (pre[0]) and adds the column title "Year". The second part of pre_clean 
# deals with each row (r) in pre from row 1 to the last row of pre. It takes each of those rows and takes the 
# item in placespot 2 (r[2]) splits it at the comma and takes the item in place number 1 [1] after the comma and strips
# it out of column in place 2 and adds it to the end of each row (r) under the newly created "Year" column for rows 
# in placespot 1 until the last row (for r in pre[1:]).

# cen uses the pivot module to create a header row based on the first row of the data. It takes the columns in the first
# row from the column in the 0 position up until the item in the 5 column and creates the header row. The titles
# "Column" and "Value" are added from the pivot module. The census.data allows for each of the first five 
# columns to be repeated for for each row for columns 2 through 37 since they were pivoted based on list(range(2,37)).

# cen_clean takes the first row of cen (cen[0]) and adds the column title "Year" to the end of it. It then goes through 
# each row of cen (r). In each row it goes to the fifth item in that row r[5], and in that fifth element it starts 
# at the 4th to last character and takes out everything but the last 4 charachters to create a new column that gets added 
# to the end of the row that only contains those for characters so that they fall under the newly added year column. 
# This is performed for rows in the 1 position to the end.
```

```python
import jupyterimporter
from csvlist import CsvList
from pivot import pivot_columns
from joiner import join_lists

preexisting = CsvList('/midterm/preexisting.csv')
census = CsvList('/midterm/census.csv')

### CLEANUP CODE STARTS
pre = pivot_columns([preexisting.header] + preexisting.data, list(range(2,37)))
cen = pivot_columns([census.header] + census.data, list(range(5,12)))

pre_clean = [pre[0] + ['Year']] + [(r + [r[2].split(',')[1].strip()]) for r in pre[1:]]
cen_clean = [cen[0] + ['Year']] + [(r + [r[5][-4:]]) for r in cen[1:]]
### CLEANUP CODE ENDS


out = join_lists(pre_clean, cen_clean, [0,4], [4,7], [0,2,3], [6])
```
